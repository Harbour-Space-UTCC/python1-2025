### set operations
union

intersection

difference (\\, -)

complementary (not) - 
$A^{'}=\mathcal{U}-A$


```python
set_a = {1, 2, 3}
set_b = {2, 3, 4}
print(set_a.union(set_b).union({4, 5, 6}))
print(set_a | set_b | {4, 5, 6})
```

```python
print(set_a.union([2, 3, 4]))
print(set_a | {2, 3, 4})
```

intersect

```python
set_a = {1, 2, 3}
set_b = {2, 3, 4}
set_a.intersection(set_b)
set_a & set_b
```

difference

```python
set_a = {1, 2, 3}
set_b = {2, 3, 4}

print(set_a.difference(set_b)) # A-B
print(set_b.difference(set_a)) # B-A
print(set_a - set_b)
print(set_b - set_a)
```

### Loop

loop by keys

```python
animal_legs = {'human': 2, 'cheetah': 4,'fish': 0, 'horse': 4, True: 6}
for key in animal_legs:
    print(key)
    print(animal_legs[key])
```

loop by .items()

```python
for item in animal_legs.items():
    print(item)
```

```python
for key, value in animal_legs.items():
    print(key)
    print(value)
```

```python
animal_legs = {'human': 2, 'cheetah': 4,'fish': 0, 'horse': 4, True: 6}
for key in animal_legs.keys():
    print(key)
```

loop by values

```python
animal_legs = {'human': 2, 'cheetah': 4,'fish': 0, 'horse': 4, True: 6}
for value in animal_legs.values():
    print(value)
```

loop by enumerate

```python
animal_legs = {'human': 2, 'cheetah': 4,'fish': 0, 'horse': 4, True: 6}
for index, element in enumerate(animal_legs):
    print(index, element)
```
