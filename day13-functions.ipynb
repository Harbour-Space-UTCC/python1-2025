```python
def square_number(x):
    """
    args:
    
    x: int - input number to be squared
    
    Square a number and returns it
    """
    result = x ** 2
    return result 
```

```python
print(square_number(5))
```

number of input(s) = $\mathbb{N} \cup \{ 0 \}$

number of output(s) = $\mathbb{N} \cup \{ 0 \}$

```python
# 0 input
# 0 output
def greet_person():
    """
    This function will print Hello 3 times
    """
    print("Hello")
    print("Hello")
    print("Hello")
    return None
```

```python
print(greet_person())
```

```python
# 1 input 1 output

def greet_person_2(name):
    """
    Greet a person according to the input name
    
    args:
    
    name (str) - name to be greeted
    """
    print(f'Hello, {name}')

print(greet_person_2("Cheetah"))
```

```python
# multiple inputs
import math
def calculate_gcd(num1, num2, num3):
    gcd_result = math.gcd(num1, num2, num3)
    return gcd_result
```

```python
print(calculate_gcd(6, 8, 9))
```

```python
# multiple outputs

def give_me_6_7():
    return 6, 7, 8
```

```python
print(type(give_me_6_7()))
```

```python
w, x, y, z = give_me_6_7()
print(x)
print(y)
print(z)
```

### *args, **kwargs

```python
def add_numbers(a, b, c, d, e):
    return a + b + c + d + e

add_numbers(4, 5, 6, 8, 10)
```

```python
def product_numbers_v2(a, b, *args, c=8):
    print(args)
    result = a * b
    for number in args:
        result *= number
    return result
    
product_numbers_v2(2, 3, 5, 6, 7, 0)
```

```python
def print_names(**kwargs):
    print(kwargs['a'])
    
print_names(x=5, y=6, z=10, a=5)
```

```python
def my_function(a, b, c=1, *args, d=5, e=8, **kwargs):
    print(a, b, c, d, args, e, kwargs)
    
my_function(1, 5, 6, 8, 10, 7, f=6, g=0, d=-1)
```

## Literal

```python
def add_numbers(a: int, b: int, c: int) -> float:
    return (a / b) / c

add_numbers("2", 5, 4)
```

```python
def double(x: float=7) -> float:
    return x * 2

double()
```

```python
def lower_all_character(word: str="DEFAULT WORD") -> None:
    return word.lower()

lower_all_character()
```
